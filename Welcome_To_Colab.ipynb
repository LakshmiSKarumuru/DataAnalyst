```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3  # Replace with your database library if necessary

!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite
# Create a connection to the database (example)
conn = sqlite3.connect('m4_survey_data.sqlite')  # Replace with your actual database

# Load your data
df = pd.read_sql_query('SELECT * FROM master', conn)

# Set up the figure with subplots
fig, axes = plt.subplots(2, 4, figsize=(25, 12))

# Histogram for ConvertedComp
sns.histplot(df['ConvertedComp'], bins=30, kde=True, color='blue', ax=axes[0, 0], alpha=0.7)
axes[0, 0].set_title('Histogram of Converted Compensation')
axes[0, 0].set_xlabel('Converted Compensation')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].grid(axis='y', alpha=0.75)

# Box plot for Age
sns.boxplot(x=df['Age'], ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title('Box Plot of Age')
axes[0, 1].set_xlabel('Age')
axes[0, 1].grid(axis='y', alpha=0.75)

# Scatter plot for Age vs WorkWeekHrs
sns.scatterplot(x='Age', y='WorkWeekHrs', data=df, ax=axes[0, 2], color='orange', alpha=0.6)
axes[0, 2].set_title('Scatter Plot of Age vs WorkWeekHrs')
axes[0, 2].set_xlabel('Age')
axes[0, 2].set_ylabel('Work Week Hours')
axes[0, 2].grid(True)

# Bubble plot for WorkWeekHrs vs CodeRevHrs with Age as bubble size
sns.scatterplot(x='WorkWeekHrs', y='CodeRevHrs', data=df, ax=axes[0, 3],
                size='Age', sizes=(20, 500), alpha=0.5, hue='Age', palette='viridis', legend=None)
axes[0, 3].set_title('Bubble Plot of WorkWeekHrs vs CodeRevHrs (Size by Age)')
axes[0, 3].set_xlabel('Work Week Hours')
axes[0, 3].set_ylabel('Code Review Hours')
axes[0, 3].grid(True)

# Pie chart for top 5 databases respondents wish to learn
top_databases = df['DatabaseWishList'].value_counts().nlargest(5)
axes[1, 0].pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140)
axes[1, 0].set_title('Top 5 Databases Respondents Wish to Learn')

# Stacked chart for median WorkWeekHrs and CodeRevHrs for age group 30 to 35
age_group = df[(df['Age'] >= 30) & (df['Age'] <= 35)]
median_values = age_group[['WorkWeekHrs', 'CodeRevHrs']].median()
median_values.plot(kind='bar', stacked=True, ax=axes[1, 1], color=['lightblue', 'salmon'])
axes[1, 1].set_title('Median WorkWeekHrs and CodeRevHrs (Age 30-35)')
axes[1, 1].set_ylabel('Hours')
axes[1, 1].set_xticklabels(['WorkWeekHrs', 'CodeRevHrs'], rotation=0)
axes[1, 1].grid(axis='y', alpha=0.75)

# Line chart for median ConvertedComp for ages 45 to 60
age_group_45_60 = df[(df['Age'] >= 45) & (df['Age'] <= 60)]
median_converted_comp = age_group_45_60.groupby('Age')['ConvertedComp'].median()
median_converted_comp.plot(ax=axes[1, 2], marker='o', color='purple')
axes[1, 2].set_title('Median ConvertedComp for Ages 45 to 60')
axes[1, 2].set_xlabel('Age')
axes[1, 2].set_ylabel('Median Converted Compensation')
axes[1, 2].grid(True)

# Horizontal bar chart for MainBranch
main_branch_counts = df['MainBranch'].value_counts()
main_branch_counts.plot(kind='barh', ax=axes[1, 3], color='skyblue')
axes[1, 3].set_title('Distribution of Main Branches')
axes[1, 3].set_xlabel('Number of Respondents')
axes[1, 3].set_ylabel('Main Branch')
axes[1, 3].grid(axis='x', alpha=0.75)

# Hide unused subplots
for ax in axes.flat:
    if not ax.has_data():
        ax.axis('off')

# Adjust layout
plt.tight_layout()
plt.show()

# Close the database connection
conn.close()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

# Load the data
try:
    !wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite
    conn = sqlite3.connect('m4_survey_data.sqlite')
    df = pd.read_sql_query('SELECT * FROM master', conn)

    # Check for the DatabaseWishList column
    if 'DatabaseWishList' not in df.columns:
        raise ValueError("Column 'DatabaseWishList' not found in the DataFrame.")

    # Count the occurrences of each database and get the top 5
    top_databases = df['DatabaseWishList'].value_counts().nlargest(5)

    if top_databases.empty:
        raise ValueError("No data available for 'DatabaseWishList'.")

    # Create a pie chart
    plt.figure(figsize=(10, 6))
    plt.pie(top_databases, labels=top_databases.index, autopct='%1.1f%%', startangle=140)
    plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
    plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the database connection if it was opened
    if 'conn' in locals():
        conn.close()
```

```
import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
!wget -q https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m4_survey_data.sqlite
conn = sqlite3.connect("m4_survey_data.sqlite")  # Open a database connection

# Query to fetch data
QUERY = """SELECT * FROM master"""
df = pd.read_sql(QUERY, conn)

# Scenario 1: Distribution of Respondents by Career Satisfaction and Job Satisfaction
pivot_table_1 = pd.pivot_table(df, index='CareerSat', columns='JobSat', values='Respondent', aggfunc='count', fill_value=0)

# Plotting for Scenario 1
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table_1, annot=True, cmap='Blues', fmt='g')
plt.title('Career Satisfaction vs. Job Satisfaction')
plt.xlabel('Job Satisfaction')
plt.ylabel('Career Satisfaction')
plt.show()

# Scenario 2: Comparison of Open Source Adoption Across Operating Systems
pivot_table_os = pd.pivot_table(df, index='OpSys', columns='OpenSource', values='Respondent', aggfunc='count', fill_value=0)

# Plotting for Scenario 2
plt.figure(figsize=(12, 6))
pivot_table_os.plot(kind='bar', width=0.8)
plt.title('Open Source Adoption Across Operating Systems')
plt.xlabel('Operating Systems')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Open Source', loc='upper right')
plt.tight_layout()  # Adjust layout for better spacing
plt.show()

# Scenario 3: Job Satisfaction and Career Aspirations by Employment Status
pivot_table_3 = pd.pivot_table(df, index=['Employment', 'JobSat'], columns='MgrWant', values='Respondent', aggfunc='count', fill_value=0)

# Plotting for Scenario 3
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table_3, annot=True, cmap='YlGnBu', fmt='g')
plt.title('Job Satisfaction and Career Aspirations by Employment Status')
plt.xlabel('Managerial Aspirations')
plt.ylabel('Employment Status and Job Satisfaction')
plt.show()

# Save the DataFrame to an Excel file
df.to_excel('./data.xlsx', index=False)

# Close the database connection
conn.close()
```
